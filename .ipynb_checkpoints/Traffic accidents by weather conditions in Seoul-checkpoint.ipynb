```python
import csv
f = open('weather_accident.csv', encoding = 'utf8')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-3] == '-':
        row[-3] = 0
    if row[-3] != '':
        result.append(int(row[-3]))
print(result)
```
