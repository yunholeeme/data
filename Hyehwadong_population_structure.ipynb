```python
import csv
f = open('age.csv')
data = csv.reader(f)

result = []
name = input("동이름 입력:")
for row in data:
    if name in row[0]:
        print(len(row[3:]))
        for i in row[3:]:
            result.append(int(i))
print(result)

import matplotlib.pyplot as plt
#print(plt.style.available)
plt.style.use('ggplot')
plt.rc('font', family='Batang')
plt.title(name + ' 지역의 인구 구조')
plt.plot(result)
plt.show()
```
