```python
import csv
f = open('age.csv')
data = csv.reader(f)

result = []
name = input("동이름 입력:")
for row in data:
    if name in row[0]:
        print(len(row[3:]))
        for i in row[3:]:
            result.append(int(i))
print(result)

import matplotlib.pyplot as plt
print(plt.style.available)
plt.style.use('ggplot')
plt.rc('font', family='Gulim')
plt.title(name)
plt.plot(result)
plt.show()
```
